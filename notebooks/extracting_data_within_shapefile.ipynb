```r
rm(list = ls())

## params ##
dirbase = "/home/jovyan"
dirdata.era5 = "https://thredds.climate.ifca.es/thredds/dodsC/fao/trainings/pakistan-202608/data/era5"

## libraries ##
library(loadeR)
#library(transformeR)
library(visualizeR)
library(sf)

## auxiliary functions ##
source(sprintf("%s/notebooks/auxiliary_functions.R", dirbase))
```

```r
## loading tmax from ERA5 (for January 2000) ##
era5 = loadGridData(sprintf("%s/tmax.nc", dirdata.era5),
                    var = "tmax", 
                    lonLim = c(60, 80), latLim = c(22, 38),
                    years = 2000, season = 1)
spatialPlot(climatology(era5), backdrop.theme = "countries")
```

```r
## extracting data within the boundaries of the shapefile ##
era5.cropped = maskGrid(era5, sprintf("%s/pk_shp/pk.shp", dirbase))
spatialPlot(climatology(era5.cropped), backdrop.theme = "countries")
```
