```r
## reading data from text files ##
library(loadeR)
data = loadStationData("/home/jovyan/obs_from_text_files", 
                       "tmax")
data$Data <- matrix(
  as.numeric(data$Data),
  nrow = nrow(data$Data),
  ncol = ncol(data$Data)
)
attr(data$Data, "dimensions") <- c("time", "loc")

```

```r
## plotting mean climatology ##
library(visualizeR)
spatialPlot(climatology(data), backdrop.theme = "countries")

## saving in .rda format ##
save(data, file = "/home/jovyan/obs_from_text_files/tmax.rda")
```
